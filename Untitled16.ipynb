````python

# -*- coding: utf-8 -*-
import os
import zipfile
import shutil
import re
import json
import requests
import time
from IPython.display import display, HTML
import ipywidgets as widgets

# ---------------------------
# إعدادات المسارات الثابتة
# ---------------------------
BASEDIR = "/content/androidproject"
ZIPNAME = "/content/FixedAndroid_Project"
TEMP_ZIP = "/content/temp_uploaded_project.zip"

# ---------------------------
# إعدادات CodeGeeX AI (محلي أو سحابي)
# ---------------------------
CODEGEEX_API_URL = "http://localhost:8080/api/generate"  # افتراضي (يمكن تغييره)
USE_CODEGEEX = True  # تفعيل/تعطيل AI

# ---------------------------
# واجهة المستخدم
# ---------------------------
style = {'description_width': 'initial'}

html_header = HTML("""
<div style="background-color: #1e1e1e; padding: 16px; border-radius: 8px; border: 1px solid #333;">
  <h3 style="color:#4CAF50; margin:0;">🤖 Android Java Project Fixer + CodeGeeX AI</h3>
  <p style="color:#bbb; margin:6px 0 0 0;">يفحص كل ملفات المشروع ويصلحها تلقائياً مع تحليل ذكي من CodeGeeX AI.</p>
</div>
""")

problem_input = widgets.Textarea(
    value="",
    placeholder="أدخل تعليمات إضافية أو سطور dependencies يدوياً (اختياري)...",
    description="📝 ملاحظات:",
    layout=widgets.Layout(width='100%', height='80px'),
    style=style
)

ai_config = widgets.HBox([
    widgets.Checkbox(value=USE_CODEGEEX, description="🤖 تفعيل CodeGeeX AI", style=style),
    widgets.Text(value=CODEGEEX_API_URL, placeholder="API URL", description="API:", layout=widgets.Layout(width='300px'), style=style)
])

uploader = widgets.FileUpload(
    accept='.zip',
    multiple=False,
    description='📤 رفع المشروع (ZIP)',
    button_style='primary'
)

run_btn = widgets.Button(
    description='🚀 بدء الفحص والإصلاح (مع AI)',
    button_style='success',
    icon='play'
)

download_btn = widgets.Button(
    description='📥 تحميل المشروع المصحح',
    button_style='info',
    icon='download',
    layout=widgets.Layout(display='none')
)

progress = widgets.IntProgress(value=0, min=0, max=100, description='📊 الإنجاز:', layout={'width': '100%'})
status = widgets.HTML(value="<b>✅ جاهز</b>")
output_area = widgets.Output()

display(html_header)
display(problem_input)
display(ai_config)
display(widgets.HBox([uploader, run_btn, download_btn]))
display(progress, status)
display(output_area)

# ---------------------------
# وظائف مساعدة
# ---------------------------
def log(msg, level="info"):
    colors = {"info": "#00d4ff", "success": "#4CAF50", "error": "#f44336", "warning": "#ff9800", "ai": "#9c27b0"}
    with output_area:
        display(HTML(f"<div style='color:{colors.get(level,'white')}; padding:4px 0;'>• {msg}</div>"))

def safe_write(path, content):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)

def read_file_safe(path):
    try:
        with open(path, 'r', encoding='utf-8', errors='ignore') as f:
            return f.read()
    except:
        return None

# ---------------------------
# CodeGeeX AI Integration
# ---------------------------
def call_codegeex_ai(prompt, file_content="", max_retries=3):
    """إرسال طلب إلى CodeGeeX API وإرجاع الرد"""
    if not USE_CODEGEEX:
        return None

    api_url = ai_config.children[1].value if ai_config.children[1].value else CODEGEEX_API_URL

    payload = {
        "prompt": f"{prompt}\n\nملف:\n```\n{file_content[:4000]}\n```" if file_content else prompt,
        "max_tokens": 2000,
        "temperature": 0.3,
        "top_p": 0.95
    }

    for attempt in range(max_retries):
        try:
            response = requests.post(api_url, json=payload, timeout=30)
            if response.status_code == 200:
                result = response.json()
                return result.get("text", result.get("response", ""))
            else:
                log(f"⚠️ خطأ API (محاولة {attempt+1}): {response.status_code}", "warning")
                time.sleep(2)
        except requests.exceptions.ConnectionError:
            log(f"⚠️ لا يمكن الاتصال بـ CodeGeeX على {api_url} (محاولة {attempt+1})", "warning")
            time.sleep(2)
        except Exception as e:
            log(f"⚠️ خطأ في الاتصال: {e} (محاولة {attempt+1})", "warning")
            time.sleep(2)

    log("❌ فشل الاتصال بـ CodeGeeX AI، سيتم استخدام الإصلاح العادي", "error")
    return None

def ai_fix_java_file(content, file_path):
    """استخدام AI لتحليل وإصلاح ملف Java"""
    prompt = f"""قم بتحليل وإصلاح ملف Java التالي. تأكد من:
1. إزالة أي BOM من بداية الملف
2. تصحيح اسم الحزمة (package) ليتوافق مع مسار المجلد
3. إضافة أي imports مفقودة (خاصة android.os.Bundle, androidx.appcompat.app.AppCompatActivity)
4. تصحيح أي أخطاء نحوية شائعة
5. تحسين الكود إن أمكن

أعد كتابة الملف كاملاً مع الإصلاحات."""

    ai_response = call_codegeex_ai(prompt, content)
    if ai_response:
        # استخراج الكود من رد الـ AI (قد يكون محاطاً بـ ```java ... ```)
        code_match = re.search(r'```(?:java)?\s*([\s\S]*?)\s*```', ai_response)
        if code_match:
            return code_match.group(1)
        return ai_response
    return None

def ai_fix_xml_file(content, file_path):
    """استخدام AI لإصلاح ملف XML"""
    prompt = f"""قم بإصلاح ملف XML التالي (Android layout/resource):
1. أضف namespaces المفقودة (android, app)
2. صحح أي أخطاء في بناء الجملة
3. تأكد من صحة المعرفات (@+id)
4. أصلح أي مشاكل في تنسيق XML

أعد كتابة الملف كاملاً."""

    ai_response = call_codegeex_ai(prompt, content)
    if ai_response:
        code_match = re.search(r'```(?:xml)?\s*([\s\S]*?)\s*```', ai_response)
        if code_match:
            return code_match.group(1)
        return ai_response
    return None

def ai_fix_manifest(content):
    """استخدام AI لإصلاح AndroidManifest.xml"""
    prompt = f"""قم بإصلاح AndroidManifest.xml التالي:
1. تأكد من وجود package صحيح
2. أضف application tag إذا مفقود
3. أضف uses-permission للإنترنت إذا كان مطلوباً
4. تأكد من وجود MainActivity مع intent-filter

أعد كتابة الملف كاملاً."""

    ai_response = call_codegeex_ai(prompt, content)
    if ai_response:
        code_match = re.search(r'```(?:xml)?\s*([\s\S]*?)\s*```', ai_response)
        if code_match:
            return code_match.group(1)
        return ai_response
    return None

def ai_fix_gradle(content, gradle_type="app"):
    """استخدام AI لإصلاح ملف Gradle"""
    prompt = f"""قم بإصلاح ملف build.gradle ({gradle_type}) التالي:
1. تأكد من وجود buildscript (لـ root) أو plugins (لـ app)
2. أضف namespace (لـ app/build.gradle)
3. أضف compileOptions مع Java 17
4. أضف dependencies الأساسية (appcompat, material, constraintlayout)

أعد كتابة الملف كاملاً."""

    ai_response = call_codegeex_ai(prompt, content)
    if ai_response:
        code_match = re.search(r'```(?:gradle)?\s*([\s\S]*?)\s*```', ai_response)
        if code_match:
            return code_match.group(1)
        return ai_response
    return None

# ---------------------------
# فحص وإصلاح ملفات Java (مع AI)
# ---------------------------
def scan_and_fix_java_files(base_dir):
    java_root = os.path.join(base_dir, "app", "src", "main", "java")
    if not os.path.exists(java_root):
        log("⚠️ مجلد Java غير موجود، سيتم إنشاؤه لاحقاً", "warning")
        return 0

    fixed_count = 0
    total_files = sum(1 for root, dirs, files in os.walk(java_root) for f in files if f.endswith(".java"))

    for i, (root, dirs, files) in enumerate(os.walk(java_root)):
        for file in files:
            if file.endswith(".java"):
                file_path = os.path.join(root, file)
                content = read_file_safe(file_path)
                if not content:
                    continue

                original_content = content

                # محاولة الإصلاح باستخدام AI أولاً
                ai_fixed = None
                if USE_CODEGEEX:
                    log(f"🤖 AI يحلل {file}...", "ai")
                    ai_fixed = ai_fix_java_file(content, file_path)
                    if ai_fixed and len(ai_fixed) > 100:
                        content = ai_fixed
                        log(f"✅ AI أصلح {file}", "success")

                # إصلاحات إضافية (احتياطية)
                if content == original_content or not ai_fixed:
                    # 1. إزالة BOM إن وجد
                    if content.startswith('\ufeff'):
                        content = content[1:]

                    # 2. إصلاح اسم الحزمة
                    rel_path = os.path.relpath(root, java_root)
                    expected_pkg = rel_path.replace(os.sep, '.') if rel_path != '.' else ''
                    if expected_pkg and expected_pkg != '.':
                        pkg_line_match = re.search(r'package\s+([\w.]+)\s*;', content)
                        if pkg_line_match and pkg_line_match.group(1) != expected_pkg:
                            content = re.sub(r'package\s+[\w.]+\s*;', f'package {expected_pkg};', content)
                            log(f"📦 إصلاح اسم الحزمة في {file}", "info")

                    # 3. إضافة imports مفقودة
                    if "import android.os.Bundle" not in content and "Bundle" in content:
                        content = re.sub(r'(package [\w.]+;)', r'\1\n\nimport android.os.Bundle;', content)

                    if "import androidx.appcompat.app.AppCompatActivity" not in content and "AppCompatActivity" in content:
                        content = content.replace("import android.os.Bundle;", "import android.os.Bundle;\nimport androidx.appcompat.app.AppCompatActivity;")

                if content != original_content:
                    safe_write(file_path, content)
                    fixed_count += 1
                    log(f"✅ إصلاح ملف Java: {file}", "success")

                progress.value = 30 + int((i + 1) / total_files * 15) if total_files > 0 else 45

    log(f"📁 تم فحص وإصلاح {fixed_count} ملف Java", "success")
    return fixed_count

# ---------------------------
# فحص وإصلاح ملفات XML (مع AI)
# ---------------------------
def scan_and_fix_xml_files(base_dir):
    res_root = os.path.join(base_dir, "app", "src", "main", "res")
    if not os.path.exists(res_root):
        log("⚠️ مجلد res غير موجود", "warning")
        return 0

    fixed_count = 0
    total_files = sum(1 for root, dirs, files in os.walk(res_root) for f in files if f.endswith(".xml"))

    for i, (root, dirs, files) in enumerate(os.walk(res_root)):
        for file in files:
            if file.endswith(".xml"):
                file_path = os.path.join(root, file)
                content = read_file_safe(file_path)
                if not content:
                    continue

                original_content = content

                # محاولة الإصلاح باستخدام AI
                if USE_CODEGEEX and "layout" in file_path:
                    ai_fixed = ai_fix_xml_file(content, file_path)
                    if ai_fixed and len(ai_fixed) > 50:
                        content = ai_fixed
                        log(f"🤖 AI أصلح {file}", "ai")

                # إصلاحات احتياطية
                if content == original_content:
                    if 'xmlns:android=' not in content:
                        content = re.sub(r'<(\w+Layout)', r'<\1 xmlns:android="http://schemas.android.com/apk/res/android"', content)

                    if 'xmlns:app=' not in content and 'xmlns:android=' in content:
                        content = content.replace('xmlns:android=', 'xmlns:android="" xmlns:app="http://schemas.android.com/apk/res-auto" ')
                        content = re.sub(r'xmlns:android=""', 'xmlns:android="http://schemas.android.com/apk/res/android"', content)

                if content != original_content:
                    safe_write(file_path, content)
                    fixed_count += 1
                    log(f"✅ إصلاح ملف XML: {file}", "success")

                progress.value = 45 + int((i + 1) / total_files * 10) if total_files > 0 else 55

    log(f"📁 تم فحص وإصلاح {fixed_count} ملف XML", "success")
    return fixed_count

# ---------------------------
# فحص وإصلاح AndroidManifest.xml (مع AI)
# ---------------------------
def scan_and_fix_manifest(base_dir):
    manifest_path = os.path.join(base_dir, "app", "src", "main", "AndroidManifest.xml")

    if not os.path.exists(manifest_path):
        os.makedirs(os.path.dirname(manifest_path), exist_ok=True)
        content = '<?xml version="1.0" encoding="utf-8"?>\n<manifest xmlns:android="http://schemas.android.com/apk/res/android">\n\n</manifest>'
    else:
        content = read_file_safe(manifest_path)

    if not content:
        content = '<?xml version="1.0" encoding="utf-8"?>\n<manifest xmlns:android="http://schemas.android.com/apk/res/android">\n\n</manifest>'

    original_content = content

    # محاولة الإصلاح باستخدام AI
    if USE_CODEGEEX:
        ai_fixed = ai_fix_manifest(content)
        if ai_fixed and len(ai_fixed) > 50:
            content = ai_fixed
            log("🤖 AI أصلح AndroidManifest.xml", "ai")

    # إصلاحات احتياطية
    if content == original_content:
        if 'package=' not in content:
            content = content.replace('<manifest', '<manifest package="com.example.app"')

        if '<application' not in content:
            content = content.replace('</manifest>', '<application android:allowBackup="true" android:label="@string/app_name"></application>\n</manifest>')

        if 'android.permission.INTERNET' not in content:
            content = content.replace('<application', '<uses-permission android:name="android.permission.INTERNET" />\n    <application')

    if content != original_content:
        safe_write(manifest_path, content)
        log("✅ تم إصلاح AndroidManifest.xml", "success")

    # استخراج package name
    pkg_match = re.search(r'package\s*=\s*"([^"]+)"', content)
    return pkg_match.group(1) if pkg_match else "com.example.app"

# ---------------------------
# فحص وإصلاح build.gradle (مع AI)
# ---------------------------
def scan_and_fix_root_build_gradle(base_dir):
    path = os.path.join(base_dir, "build.gradle")
    content = read_file_safe(path) if os.path.exists(path) else ""

    if not content:
        content = """buildscript {
    repositories {
        google()
        mavenCentral()
    }
    dependencies {
        classpath 'com.android.tools.build:gradle:8.2.0'
    }
}

allprojects {
    repositories {
        google()
        mavenCentral()
    }
}
"""
    else:
        original_content = content
        if USE_CODEGEEX:
            ai_fixed = ai_fix_gradle(content, "root")
            if ai_fixed and len(ai_fixed) > 50:
                content = ai_fixed
                log("🤖 AI أصلح build.gradle (root)", "ai")

        if content == original_content:
            if 'buildscript' not in content:
                content = """buildscript {
    repositories {
        google()
        mavenCentral()
    }
    dependencies {
        classpath 'com.android.tools.build:gradle:8.2.0'
    }
}\n\n""" + content
            if 'google()' not in content:
                content = content.replace('repositories {', 'repositories {\n        google()')

    safe_write(path, content)
    log("✅ build.gradle (root) جاهز", "success")

def scan_and_fix_app_build_gradle(base_dir, package_name):
    app_dir = os.path.join(base_dir, "app")
    os.makedirs(app_dir, exist_ok=True)
    path = os.path.join(app_dir, "build.gradle")

    content = read_file_safe(path) if os.path.exists(path) else ""

    if not content:
        content = f"""plugins {{
    id 'com.android.application'
}}

android {{
    namespace "{package_name}"
    compileSdk 34

    defaultConfig {{
        applicationId "{package_name}"
        minSdk 24
        targetSdk 34
        versionCode 1
        versionName "1.0"
    }}

    compileOptions {{
        sourceCompatibility JavaVersion.VERSION_17
        targetCompatibility JavaVersion.VERSION_17
    }}
}}

dependencies {{
    implementation 'androidx.appcompat:appcompat:1.6.1'
    implementation 'com.google.android.material:material:1.11.0'
    implementation 'androidx.constraintlayout:constraintlayout:2.1.4'
}}
"""
    else:
        original_content = content
        if USE_CODEGEEX:
            ai_fixed = ai_fix_gradle(content, "app")
            if ai_fixed and len(ai_fixed) > 50:
                content = ai_fixed
                log("🤖 AI أصلح app/build.gradle", "ai")

        if content == original_content:
            if 'namespace' not in content:
                content = re.sub(r'android\s*\{', f'android {{\n    namespace "{package_name}"', content)
            if 'compileOptions' not in content:
                content = content.replace('android {', f'android {{\n    compileOptions {{\n        sourceCompatibility JavaVersion.VERSION_17\n        targetCompatibility JavaVersion.VERSION_17\n    }}\n')
            if 'dependencies' not in content:
                content += """

dependencies {
    implementation 'androidx.appcompat:appcompat:1.6.1'
    implementation 'com.google.android.material:material:1.11.0'
    implementation 'androidx.constraintlayout:constraintlayout:2.1.4'
}
"""

    safe_write(path, content)
    log("✅ app/build.gradle جاهز", "success")

# ---------------------------
# باقي وظائف الإصلاح الأساسية
# ---------------------------
def scan_and_fix_settings_gradle(base_dir):
    path = os.path.join(base_dir, "settings.gradle")
    content = read_file_safe(path) if os.path.exists(path) else ""

    if not content:
        content = "rootProject.name = 'MyApplication'\ninclude ':app'\n"
    elif "include ':app'" not in content:
        content += "\ninclude ':app'\n"

    safe_write(path, content)
    log("✅ settings.gradle جاهز", "success")

def scan_and_fix_gradle_wrapper(base_dir):
    wrapper_dir = os.path.join(base_dir, "gradle", "wrapper")
    os.makedirs(wrapper_dir, exist_ok=True)
    path = os.path.join(wrapper_dir, "gradle-wrapper.properties")

    expected_url = "distributionUrl=https\\://services.gradle.org/distributions/gradle-8.2-bin.zip"
    safe_write(path, expected_url + "\n")
    log("✅ gradle-wrapper.properties جاهز", "success")

def scan_and_fix_gradle_properties(base_dir):
    path = os.path.join(base_dir, "gradle.properties")
    content = """org.gradle.jvmargs=-Xmx2g
android.useAndroidX=true
android.enableJetifier=true
"""
    safe_write(path, content)
    log("✅ gradle.properties جاهز", "success")

def ensure_main_activity_exists(base_dir, package_name):
    java_root = os.path.join(base_dir, "app", "src", "main", "java")
    pkg_path = os.path.join(java_root, *package_name.split("."))
    os.makedirs(pkg_path, exist_ok=True)

    main_activity_path = os.path.join(pkg_path, "MainActivity.java")

    if not os.path.exists(main_activity_path):
        content = f"""package {package_name};

import android.os.Bundle;
import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {{
    @Override
    protected void onCreate(Bundle savedInstanceState) {{
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);
    }}
}}
"""
        safe_write(main_activity_path, content)
        log("➕ إنشاء MainActivity.java جديد", "success")

def ensure_basic_resources(base_dir):
    layout_dir = os.path.join(base_dir, "app", "src", "main", "res", "layout")
    os.makedirs(layout_dir, exist_ok=True)

    main_layout = os.path.join(layout_dir, "activity_main.xml")
    if not os.path.exists(main_layout):
        content = """<?xml version="1.0" encoding="utf-8"?>
<androidx.constraintlayout.widget.ConstraintLayout xmlns:android="http://schemas.android.com/apk/res/android"
    xmlns:app="http://schemas.android.com/apk/res-auto"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <TextView
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"
        android:text="Hello World!"
        app:layout_constraintBottom_toBottomOf="parent"
        app:layout_constraintLeft_toLeftOf="parent"
        app:layout_constraintRight_toRightOf="parent"
        app:layout_constraintTop_toTopOf="parent" />

</androidx.constraintlayout.widget.ConstraintLayout>
"""
        safe_write(main_layout, content)
        log("➕ إنشاء activity_main.xml جديد", "success")

    values_dir = os.path.join(base_dir, "app", "src", "main", "res", "values")
    os.makedirs(values_dir, exist_ok=True)

    strings_path = os.path.join(values_dir, "strings.xml")
    if not os.path.exists(strings_path):
        content = """<?xml version="1.0" encoding="utf-8"?>
<resources>
    <string name="app_name">MyApp</string>
</resources>
"""
        safe_write(strings_path, content)
        log("➕ إنشاء strings.xml جديد", "success")

def add_manual_dependencies(base_dir, notes):
    if not notes:
        return

    app_gradle_path = os.path.join(base_dir, "app", "build.gradle")
    if not os.path.exists(app_gradle_path):
        return

    content = read_file_safe(app_gradle_path)
    if not content:
        return

    additions = []
    for line in notes.splitlines():
        line = line.strip()
        if not line or line.startswith('//'):
            continue
        if line.startswith(("implementation", "api", "compileOnly")):
            additions.append(line)
        elif ':' in line:
            additions.append("implementation '" + line.strip().strip("'\"") + "'")

    if not additions:
        return

    if 'dependencies {' not in content:
        content += "\n\ndependencies {\n}\n"

    def add_deps(txt):
        match = re.search(r'(dependencies\s*\{)([\s\S]*?)(\n\})', txt)
        if match:
            return txt[:match.end(1)] + match.group(2) + "\n    " + "\n    ".join(additions) + match.group(3)
        return txt

    new_content = add_deps(content)
    if new_content != content:
        safe_write(app_gradle_path, new_content)
        log(f"➕ إضافة {len(additions)} dependency يدوياً", "info")

# ---------------------------
# الفحص الرئيسي
# ---------------------------
def full_project_scan_and_fix(base_dir, user_notes=""):
    global USE_CODEGEEX
    USE_CODEGEEX = ai_config.children[0].value

    log("🔍 بدء الفحص الكامل للمشروع..." + (" (مع CodeGeeX AI)" if USE_CODEGEEX else ""), "info")

    progress.value = 20
    scan_and_fix_java_files(base_dir)

    progress.value = 40
    scan_and_fix_xml_files(base_dir)

    progress.value = 55
    package_name = scan_and_fix_manifest(base_dir)
    log(f"📦 اسم الحزمة: {package_name}", "info")

    progress.value = 65
    scan_and_fix_root_build_gradle(base_dir)
    scan_and_fix_app_build_gradle(base_dir, package_name)
    scan_and_fix_settings_gradle(base_dir)
    scan_and_fix_gradle_properties(base_dir)
    scan_and_fix_gradle_wrapper(base_dir)

    if user_notes:
        add_manual_dependencies(base_dir, user_notes)

    progress.value = 85
    ensure_main_activity_exists(base_dir, package_name)
    ensure_basic_resources(base_dir)

    progress.value = 95
    log("🎉 اكتمل الفحص والإصلاح بنجاح!", "success")

# ---------------------------
# معالجة رفع وتحميل المشروع
# ---------------------------
def on_run_clicked(b):
    if not uploader.value:
        status.value = "<b style='color:red'>❌ يرجى رفع ملف ZIP أولاً!</b>"
        return

    with output_area:
        output_area.clear_output()

    progress.value = 5
    status.value = "⏳ جاري المعالجة..."
    download_btn.layout.display = 'none'

    try:
        if os.path.exists(BASEDIR):
            shutil.rmtree(BASEDIR)
        if os.path.exists(TEMP_ZIP):
            os.remove(TEMP_ZIP)

        uploaded = list(uploader.value.values())[0]
        with open(TEMP_ZIP, "wb") as f:
            f.write(uploaded['content'])

        progress.value = 10
        log("✅ تم حفظ ملف ZIP", "success")

        with zipfile.ZipFile(TEMP_ZIP, 'r') as z:
            z.extractall(BASEDIR)

        progress.value = 15
        log("✅ تم فك الضغط", "success")

        notes = problem_input.value.strip()
        full_project_scan_and_fix(BASEDIR, user_notes=notes)

        progress.value = 100
        status.value = "<b style='color:green'>✅ تم الفحص والإصلاح بنجاح!</b>"
        download_btn.layout.display = 'inline-block'

        final_zip = ZIPNAME + ".zip"
        if os.path.exists(final_zip):
            os.remove(final_zip)
        shutil.make_archive(ZIPNAME, 'zip', BASEDIR)

        log("📦 المشروع المصحح جاهز للتحميل", "success")

    except Exception as e:
        log(f"❌ خطأ: {e}", "error")
        status.value = "<b style='color:red'>❌ حدث خطأ أثناء المعالجة</b>"

def on_download_clicked(b):
    try:
        from google.colab import files
        final_path = ZIPNAME + ".zip"
        if os.path.exists(final_path):
            files.download(final_path)
        else:
            log("الملف غير جاهز", "error")
    except Exception as e:
        log(f"فشل التحميل: {e}", "error")

run_btn.on_click(on_run_clicked)
download_btn.on_click(on_download_clicked)
````
